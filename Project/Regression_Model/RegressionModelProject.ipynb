import all the libraries we need 

```python
import tensorflow as tf
from keras import layers
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Load data from dataset

```python
ds = pd.read_csv("https://cocl.us/concrete_data")
ds.head()
```

Check if there are any null value

```python
ds.isnull().sum()
```

```python
class_name = ds.columns
class_name
```

Separate predictors and target

```python
x_train = ds[class_name[class_name != "Strength"]]
y_train = ds[class_name[class_name == "Strength"]]
print(x_train.head())
print(y_train.head())
```

Normalize the predictor variables

```python
x_train = (x_train-x_train.mean())/x_train.std()
x_train.head()
```

Split train and test dataset

```python
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.3)
```

Build the model

```python
model = tf.keras.Sequential([
    layers.Dense(10, activation="relu", input_shape = (x_train.shape[1],)),
    layers.Dense(10, activation="relu"),
    layers.Dense(10, activation="relu"),
    layers.Dense(1)
])
model.summary()
model.compile(optimizer="adam", loss = "mean_squared_error")
```

Train the model

```python
history = model.fit(x_train, y_train, validation_split=0.3, epochs=50, verbose=2)
```

```python
predicted = model.predict(x_test)
mse = metrics.mean_squared_error(predicted, y_test)
mse
```

Repeat the above step and generate a general function for the regression process

```python
def regressionModel():
    ds = pd.read_csv("https://cocl.us/concrete_data")
    class_name = ds.columns
    x_train = ds[class_name[class_name != "Strength"]]
    y_train = ds[class_name[class_name == "Strength"]]
    x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.3)
    model = tf.keras.Sequential([
        layers.Dense(10, activation="relu", input_shape = (x_train.shape[1],)),
        layers.Dense(10, activation="relu"),
        layers.Dense(10, activation="relu"),
        layers.Dense(1)
    ])
    model.compile(optimizer="adam", loss = "mean_squared_error")
    model.fit(x_train, y_train, validation_split=0.3, epochs=50, verbose=2)
    predicted = model.predict(x_test)
    mse = metrics.mean_squared_error(y_test, predicted)
    return mse
```

```python
mseList = []
for i in range(50):
    mseList.append(regressionModel())

mean = np.mean(mseList)
std = np.std(mseList)
print(f"mean: {mean} \nstd: {std}")
```
