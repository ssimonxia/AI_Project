```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./train.csv")
df
```

```python
print(len(df))
print(df.shape)
```

```python
df.columns
```

```python
for i in df.columns:
    print(df[i].dtype)
```

```python
dataset_X = df[df.columns[:len(df.columns) - 1]]
dataset_Y = df["price_range"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset_X, dataset_Y, test_size=0.3, random_state=111)
```

```python
dtree = DecisionTreeClassifier(criterion="entropy", max_depth = 8)
dtree
```

```python
dtree.fit(X_train, y_train)
```

```python
yhat = dtree.predict(X_test)
```

```python
correct = 0
for i, j in zip(yhat, y_test):
    if i == j:
        correct += 1
correct/len(yhat)
```
